```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

```python
#Almacena la informacion de la columna "Precio" de los archivos csv llamados "tienda"
ingresos_por_tienda = {
    'tienda': tienda['Precio'].sum(),
    'tienda2': tienda2['Precio'].sum(),
    'tienda3': tienda3['Precio'].sum(),
    'tienda4': tienda4['Precio'].sum()
}
#Calcular ingreso total, sumando la columna precio de cada conjunto de tienda
def ingreso_total():
  return sum(ingresos_por_tienda.values())
```

```python
#Organizo los ingresos de cada tienda y el total en una lista
ingresos = {
    "Tienda": ["Tienda", "Tienda 2", "Tienda 3", "Tienda 4","Total"],
    "Ingreso": [tienda['Precio'].sum(),
                tienda2['Precio'].sum(),
                tienda3['Precio'].sum(),
                tienda4['Precio'].sum(),
                ingreso_total()
                ]
}

tabla_ingresos = pd.DataFrame(ingresos)
tabla_ingresos.head()
```
